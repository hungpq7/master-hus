# 3.21. Vũ Thị Kim Trâm

Cho tập dữ liệu paper (bài 3.13) đo lường độ bền của giấy Kraft và hàm lượng gỗ cứng trong giấy.

1. Vẽ biểu đồ để  cho thấy rằng dữ liệu có mối quan hệ phi tuyến tính.
2. Xác định một mô hình đa thức (polynomial) phù hợp cho dữ liệu bằng cách sử dụng hàm poly().
3. Xác định một mô hình phù hợp bằng cách sử dụng regression spline.
4. Vẽ hai mô hình (một sử dụng poly(), một sử dụng regression spline) trên dữ liệu gốc và đưa ra nhận xét.

```r
# 3.21. Tạo bảng dữ liệu paper từ thông tin trong bài 3.13.

paper <- data.frame(
  strength = c(6.3, 11.1, 20.0, 24.0, 26.1, 30.0, 33.8, 34.0, 38.1, 39.9, 42.0, 46.1, 53.1, 52.0, 52.5, 48.0, 42.8, 27.8, 21.9),
  hardwood = c(1.0, 1.5, 2.0, 3.0, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 8.0, 9.0, 10.0, 11.0, 12.0, 13.0, 14.0, 15.0)
); head(paper)
```

```r
# 3.21.1 Vẽ biểu đồ, thể hiện rằng dữ liệu không có quan hệ tuyến tính
plot( strength ~ hardwood ,
  data = paper,
  ylab = "Strength",
  xlab = "Hardwood",
  pch = 19)
```

Ta thấy biểu đồ kết quả không phải là quan hệ tuyến tính đơn giản, có thể làm hàm bậc 2 trở lên

```r
# 3.21.2. Xác định một mô hình đa phức phù hợp, sử dụng poly()

# Xây dựng các mô hình bậc 1 đến bậc 4
model_poly_1 <- lm(strength ~ poly(hardwood, 1), data = paper)
model_poly_2 <- lm(strength ~ poly(hardwood, 2), data = paper)
model_poly_3 <- lm(strength ~ poly(hardwood, 3), data = paper)
model_poly_4 <- lm(strength ~ poly(hardwood, 4), data = paper)

summary(model_poly_1)
summary(model_poly_2)
summary(model_poly_3)
summary(model_poly_4)
```

```r
# So sánh các mô hình sử dụng
AIC(model_poly_1, model_poly_2, model_poly_3, model_poly_4)
BIC(model_poly_1, model_poly_2, model_poly_3, model_poly_4)
```

Lựa chọn mô hình số 3 vì có AIC và BIC thấp nhất


```r
# @title
# Biểu diễn trực quan 4 mô hình bậc 1-4

# Thiết lập bố cục trang với 4 phần
par(mfrow = c(2, 2)) # 2 hàng, 2 cột

# Vẽ biểu đồ cho từng mô hình
plot(strength ~ hardwood, main = "Polynomial Degree 1", data = paper,
     xlab = "Hardwood", ylab = "Strength", pch = 19)
lines(paper$hardwood, predict(model_poly_1, newdata=paper), lwd=2)

plot(strength ~ hardwood, main = "Polynomial Degree 2", data = paper,
     xlab = "Hardwood", ylab = "Strength", pch = 19)
lines(paper$hardwood, predict(model_poly_2, newdata=paper), lwd=2)

plot(strength ~ hardwood, main = "Polynomial Degree 3", data = paper,
     xlab = "Hardwood", ylab = "Strength", pch = 19)
lines(paper$hardwood, predict(model_poly_3, newdata=paper), lwd=2)

plot(strength ~ hardwood, main = "Polynomial Degree 4", data = paper,
     xlab = "Hardwood", ylab = "Strength", pch = 19)
lines(paper$hardwood, predict(model_poly_4, newdata=paper), lwd=2)

```

```r
par(mfrow = c(2,2))
plot(model_poly_3, which = c(1,2,4,5), main = "Model Poly 3")
```

```r
# 3.21.3. Xác định một mô hình phù hợp sử dụng hồi quy spline

# Tải thư viện
library(splines)

# Thử nghiệm với các mô hình spline có df - số bậc tự do khác nhau
model_spline_df3 <- lm(strength ~ bs(hardwood, df = 3), data = paper)
model_spline_df4 <- lm(strength ~ bs(hardwood, df = 4), data = paper)
model_spline_df5 <- lm(strength ~ bs(hardwood, df = 5), data = paper)
model_spline_df6 <- lm(strength ~ bs(hardwood, df = 6), data = paper)

# Lựa chọn mô hình phù hợp bằng chỉ số AIC và BIC

AIC(model_spline_df3, model_spline_df4, model_spline_df5, model_spline_df6)
BIC(model_spline_df3, model_spline_df4, model_spline_df5, model_spline_df6)
```

AIC và BIC càng thấp càng tốt. Do đó dựa chon mô hình model_spline_df5

```r
# @title

plot(strength ~ hardwood, main="Spline Model: df=3", data = paper,
  xlab="Hardwood", ylab="Strength", pch=19)
lines(paper$hardwood, predict(model_spline_df3))

plot(strength ~ hardwood, main="Spline Model: df=4", data = paper,
  xlab="Hardwood", ylab="Strength", pch=19)
lines(paper$hardwood, predict(model_spline_df4))

plot(strength ~ hardwood, main="Spline Model: df=5", data = paper,
  xlab="Hardwood", ylab="Strength", pch=19)
lines(paper$hardwood, predict(model_spline_df5))

plot(strength ~ hardwood, main="Spline Model: df=6", data = paper,
  xlab="Hardwood", ylab="Strength", pch=19)
lines(paper$hardwood, predict(model_spline_df6))
```

```r
par(mfrow = c(2,2))
plot(model_spline_df5, which = c(1,2), main = "Model Spline 5")
plot(model_spline_df5, which = c(1,2), main = "Model Spline 6")
```

```r
# 3.21.4. Vẽ 2 mô hình được lựa chọn theo poly() và theo spline() trên dữ liệu gốc và so sánh

# Chia màn hình về 1 phần
par(mfrow=c(1, 1))

# 1. Vẽ biểu đồ dữ liệu gốc (scatter plot)
plot(strength ~ hardwood, main="Data with Polynomial and Spline Models", data = paper,
     xlab="Hardwood", ylab="Strength", pch=19)

# Vẽ đường mô hình polynomial bậc 3
lines(paper$hardwood, predict(model_poly_3, newdata=paper), col="blue", lwd=2)

# Vẽ đường mô hình spline df =.
lines(model_spline_df5, col="red", lwd=2)

# 4. Thêm chú thích để phân biệt hai mô hình
legend("topleft", legend=c("Polynomial (degree 3)", "Spline (df=5)"),
       col=c("blue", "red"), lty=1, lwd=2)
```

# 3.22. Triệu Thị Như Quỳnh

3.22. Thí nghiệm đã được thực hiện để đo nhiệt lượng sinh ra trong quá trình xi măng đóng rắn với các thành phần thay đổi.

1. Vẽ biểu đồ từng biến giải thích so với nhiệt lượng sinh ra và đưa ra nhận định thành phần nào có liên quan đến nhiệt lượng sinh ra.
2. Ước lượng mô hình hồi quy tuyến tính dự đoán nhiệt lượng sinh ra từ các biến giải thích A, B, C và D (không xét tương tác). Sử dụng kiểm định t, xác định các biến giải thích nào có ý nghĩa thống kê. So sánh với nhận định ở phần trước.
3. Chứng minh rằng đa cộng tuyến có thể là một vấn đề. Giải thích lý do tại sao điều này có thể xảy ra và đề xuất một giải pháp.
4. Ước lượng mô hình đã chỉnh sửa và so sánh kết quả kiểm định t với kết quả kiểm định t từ mô hình ban đầu.

```r
# Nhập lại dữ liệu
data_heat <- data.frame(
  A = c(7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10),
  B = c(26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68),
  C = c(6, 15, 8, 8, 6, 9, 17, 22, 12, 4, 23, 9, 8),
  D = c(60, 52, 20, 47, 33, 22, 6, 44, 28, 26, 34, 12, 12),
  Heat = c(78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4)
  )
print(data_heat)
```

Giải thích biến:
- **Heat:** Nhiệt lượng toả ra.
- **A:** Tricalcium aluminate (viết tắt là C₃A) là một hợp chất hóa học có công thức hóa học là Ca₃Al₂O₆
- **B**: Tricalcium silicate (viết tắt là C₃S) là một hợp chất hóa học có công thức hóa học là Ca₃SiO₅.
- **C**: Alcium alumino ferrite (viết tắt là C₄AF) là một hợp chất hóa học có công thức hóa học là Ca₄Al₂Fe₂O₁₀.
- **D**: Dicalcium silicate (viết tắt là C₂S) có công thức hóa học là Ca₂SiO₄


```r
#1 Vẽ biểu đồ
par(mfrow=c(2, 2))

plot(A ~ Heat, main="Heat and A", data = data_heat,
     xlab="A", ylab="Heat", pch=19)
plot(B ~ Heat, main="Heat and B", data = data_heat,
     xlab="B", ylab="Heat", pch=19)
plot(C ~ Heat, main="Heat and C", data = data_heat,
     xlab="C", ylab="Heat", pch=19)
plot(D ~ Heat, main="Heat and D", data = data_heat,
     xlab="D", ylab="Heat", pch=19)
```

**Từ biểu đồ trên ta thấy:**
*   Khi A tăng Heat có điểm tăng có điểm giảm một cách ngẫu nhiên => A không có mối quan hệ rõ ràng với Heat.
*   Khi B tăng, Heat cũng tăng. Điểm dữ liệu gần như nằm trên một đường xu hướng tăng => B có mối quan hệ tuyến tính mạnh với Heat, đây có thể là biến quan trọng.
*   Khi C tăng Heat thay đổi không theo quy luật (tăng giảm ngẫu nhiên) => C không có mối quan hệ với Heat
*   Khi D tăng , Heat có xu hướng giảm nhưng các điểm dữ liệu không tập trung => D có mối quan hệ không quá mạnh với Heat





```r
#2 Hồi quy tuyến tính
model <- lm(Heat ~ A + B + C + D, data = data_heat)
summary(model)
```

- Tất cả các biến đều có p-value > 0.05 do đó các hệ số đều không có ý nghĩa thống kê.
- P-value của A là 0.109 nên có thể nói A có ảnh hưởng tới Heat nhưng không đủ mạnh để bác bỏ H0.
- Nếu chỉ dựa vào kết quả hồi quy hiện tại thì nhận định về mối quan hệ giữa các biến với Heat trong phần trước đang không được chính xác.

**Kiểm định t (t-test)**
Kiểm định t trong hồi quy tuyến tính được sử dụng để kiểm tra xem các hệ số hồi quy có khác biệt có ý nghĩa thống kê so với 0 hay không. Cụ thể, đối với mỗi biến giải thích, kiểm định t kiểm tra giả thuyết:
*   H0: Hệ số hồi quy của biến giải thích bằng 0
*   H1: Hệ số hồi quy của biến giải thích đó khác 0.
Giả thuyết được kiểm tra bằng cách sử dụng giá trị t-value và p-value.
Mỗi hệ số hồi quy có một p-value, cho biết khả năng xảy ra của t-value dưới giả thuyết H0. Nếu p-value nhỏ hơn 0.05 (hoặc mức ý nghĩa nào đó), chúng ta bác bỏ giả thuyết H0.

```r
#3 Tính toán chỉ số VIF
install.packages("car")
library(car)
vif(model)
```

**Nhận xét:**
*   Tất cả các biến (A, B, C, D) đều có VIF rất cao, đặc biệt là B và D với VIF lần lượt là 252.91 và 275.25. Điều này cho thấy mức độ đa cộng tuyến nghiêm trọng trong mô hình.

Lựa chọn cách loại bỏ các biến có tương quan cao để xây dựng lại mô hình này.

**Đa cộng tuyến (Multicollinearity)** xảy ra khi có sự tương quan cao giữa các biến giải thích trong mô hình hồi quy tuyến tính. Điều này có thể làm cho các hệ số hồi quy trở nên không ổn định và không đáng tin cậy, gây khó khăn trong việc xác định tác động của từng biến giải thích đối với biến phụ thuộc.

**Đo lường đa cộng tuyến với VIF**

Để xác định xem đa cộng tuyến có phải là vấn đề trong mô hình hay không, chúng ta có thể tính toán Variance Inflation Factor (VIF) cho từng biến giải thích. VIF đo lường mức độ mà sự biến động của một biến giải thích bị "phồng lên" do sự tương quan với các biến giải thích khác. Nếu VIF quá cao, điều này cho thấy sự tồn tại của đa cộng tuyến.

VIF lớn hơn 5: Cho thấy có tồn tại đa cộng tuyến.

VIF nhỏ hơn hoặc bằng 5: Mối quan hệ giữa các biến giải thích không phải là vấn đề lớn.

**Ngưỡng để đánh giá mức độ tương quan:**

- 0.3 ≤ ∣r∣ < 0.5: Tương quan trung bình.
- 0.5 ≤ ∣r∣ < 0.7: Tương quan khá mạnh.
- 0.7 ≤ ∣r∣ < 0.9: Tương quan mạnh.
- ∣r∣ ≥ 0.9: Tương quan rất mạnh (gần như hoàn hảo).

**2. Giải thích tại sao mô hình hồi quy tuyến tính có trọng số là phù hợp.**

Mô hình hồi quy tuyến tính có trọng số (weighted linear regression) là một dạng của hồi quy tuyến tính, trong đó mỗi điểm dữ liệu có một trọng số riêng, phản ánh độ tin cậy hoặc tầm quan trọng của mỗi quan sát trong mô hình.

**Mô hình hồi quy có trọng số phù hợp trong bài này vì:**<br>

**Sự khác biệt về kích thước mẫu:**<br>
Các nghiên cứu trong bộ dữ liệu có SampleSize khác nhau, tức là số lượng quan sát có thể khác nhau giữa các nghiên cứu. Kích thước mẫu lớn hơn có thể cung cấp thông tin đáng tin cậy hơn về mối quan hệ giữa các biến. Do đó, trong mô hình hồi quy, các quan sát với SampleSize lớn sẽ được ưu tiên hơn, nên có trọng số lớn hơn

**Giảm sai số ước lượng:**<br>

5. Ước lượng tham số mô hình dạng y^2 ~ I(x1^2) + I(x2^2) - 1. Sử dụng t-test để xác nhận ước lượng tham số được xác định bởi định lí Pytago cũng được xác nhận trên dữ liệu

```r
# Khoi tao dataset
dataset <- data.frame(
    dataset <- data.frame(
        y = numeric(20),
        x1 = numeric(20),
        x2 = numeric(20)
    )
)

dataset$y <- c(10.1, 11.6, 10.4, 13.0, 12.3, 11.1, 11.4, 13.0, 13.2, 10.2, 8.8, 13.5, 10.3, 12.6, 10.1, 10.9, 12.2, 11.3, 10.1, 9.7)
dataset$x1 <- c(5.3, 5.4, 4.5, 4.7, 6.6, 4.2, 5.0, 5.0, 6.9, 4.7, 4.2, 5.6, 3.2, 6.5, 4.3, 5.7, 4.0, 4.2, 5.6, 5.6)
dataset$x2 <- c(8.5, 10.3, 9.4, 12.2, 10.4, 10.3, 10.2, 12.1, 11.2, 9.0, 7.7, 12.3, 9.8, 10.8, 9.1, 9.3, 11.6, 10.4, 8.5, 7.9)


# Fit linear regression model
model <- lm(y ~ x1 + x2, data = dataset)

# Print the model summary
summary(model)
```

1. Ở đây, hệ số chặn (Intercept) có p-value > 0.05. Vì thế không thể bác bỏ giả thiết Ho : Intercept = 0. Nói cách khác, hệ số chặn không cần thiết trong mô hình

```r
# Su dung QQ plot de kiem tra tinh phan phoi chuan cua phan du
qqnorm(model$residuals)
qqline(model$residuals)

# Su dung bai kiem dinh sharpiro. H0: phần dư tuân theo phân phối chuẩn
shapiro.test(model$residuals)
```

```r
# Kiểm định giả thuyết phương sai sai số thay đổi
plot(model$fitted.values, model$residuals)

# H0: phương sai sai số không đổi
bp_test <- bptest(model)
print(bp_test)

# Plot residual against its lag 1
plot(model$residuals[-1], model$residuals[-length(model$residuals)])

# Durbin-Watson test
# install.packages("lmtest")
library(lmtest)

# H0: Không tồn tại tự tương quan
dwtest(model)
```

*   Biểu đồ phần dư - giá trị ước lượng y cho thấy phương sai sai số không có mối quan hệ phụ thuộc với giá trị y. Bài test Breusch-Pagan cũng cho kết quả tương tự (p-value = 0.77 -> không bác bỏ giả thuyết H0: phương sai sai số không đổi)

Tuy nhiên, trên đồ thị residuals vs leverage không có điểm nào nằm ra khỏi đường bào Cook's Distance (đường gạch đứt) nên có thể kết luận không có điểm dữ liệu nào quá bất thường.


```r
model1 <- lm(y ~ x1 + x2 - 1, data = dataset)
summary(model1)
```

3. Diễn giải mô hình: Mô hình biểu thị mối quan hệ tuyến tính giữa biến phụ thuộc y và 2 biến độc lập x1, x2. Hệ số của x1 là 0.462161 và hệ số của x2 là 0.890454. Mô hình có các hệ số có ý nghĩa thống kê, giá trị R-squared điều chỉnh bằng 1.

4. Việc mô hình tuyến tính vẫn ước lượng được trong khi mô hình dữ liệu thực tế có dạng y = sqrt(x1^2 + x2^2) có nguyên nhân đến từ việc việc chọn mẫu (sampling method). Thực tế giá trị mẫu của x1 va x2 quá nhỏ so với miền giá trị của toàn bộ tổng thế.
Nếu biểu diễn mẫu trên không gian 3 chiều thì các điểm sẽ gần như năm trên 1 mặt phẳng, tuy nhiên toàn bộ không gian mẫu lại là 1 dạng hình nón.

```r
# Fit the model for μ = E[y^2] using the systematic component I(x1^2) + I(x2^2) - 1
model_mu <- lm(I(y^2) ~ I(x1^2) + I(x2^2) - 1, data = dataset)

# Print the model summary
summary(model_mu)
```

```r
# Tính toán t-statistic cho giả thuyết H0: coef = 1

# Calculate the coefficients minus 1
coefficients_minus_1 <- coef(model_mu) - 1

# Calculate the standard errors
std_errors <- summary(model_mu)$coefficients[, "Std. Error"]

# Divide the coefficients minus 1 by their standard errors
result <- coefficients_minus_1 / std_errors

# Print the result
print(result)
```

```r
# 2. Check variance and transformations
# Residual variance assessment
par(mfrow = c(1, 1))
model_raw <- lm(Butterfat ~ Breed + Age, data = data)
plot(model_raw, which = 1)  # Residuals vs Fitted

# Apply transformations
data$Log_Butterfat <- log(data$Butterfat)
data$Sqrt_Butterfat <- sqrt(data$Butterfat)
data$Inv_Butterfat <- 1 / data$Butterfat
data$BoxCox_Butterfat <- (data$Butterfat^(-1.4) - 1) / -1.4

# Fit models with transformed response
model_log <- lm(Log_Butterfat ~ Breed + Age, data = data)
model_sqrt <- lm(Sqrt_Butterfat ~ Breed + Age, data = data)
model_inv <- lm(Inv_Butterfat ~ Breed + Age, data = data)
model_coxbox <- lm(BoxCox_Butterfat ~ Breed + Age, data = data)

# Plot residuals for transformed models
par(mfrow = c(2, 2))
scatter.smooth(rstandard(model_sqrt) ~ fitted(model_sqrt), las = 1, col = "grey",
                ylab = "Standardized residuals", xlab = "Fitted values",
                main = "Square-root Transformation")
scatter.smooth(rstandard(model_log) ~ fitted(model_log), las = 1, col = "grey",
                ylab = "Standardized residuals", xlab = "Fitted values",
                main = "Log Transformation")
scatter.smooth(rstandard(model_inv) ~ fitted(model_inv), las = 1, col = "grey",
                ylab = "Standardized residuals", xlab = "Fitted values",
                main = "Inverse Transformation")
scatter.smooth(rstandard(model_coxbox) ~ fitted(model_coxbox), las = 1, col = "grey",
                ylab = "Standardized residuals", xlab = "Fitted values",
                main = "Cox-box 1.4 Transformation")

# Box-Cox transformation
par(mfrow = c(1, 1))
boxcox(model_raw, lambda = seq(-2, 2, by = 0.1))
```

- Đồ thị phần dư khi áp dụng log-transform có xu hướng dao động ít khi y thấp nhưng dao động nhiều hơn khi y tăng.
- Đồ thị phần dư khi áp dụng các phương pháp inverse transform có xu hướng tương đối ổn định.
- Quan sát đồ thị box-cox, khi áp dụng transform với $\lambda = 1.4$ thì đồ thị phần dư giữ nguyên nhất. Đồ thị box-cox giúp đưa ra lựa chọn này.

```r
# Load necessary libraries
library(car)      # For additional regression diagnostics
library(lmtest)   # For testing autocorrelation
library(nortest)  # For testing normality of residuals

# Fit the regression model
model <- lm(BoxCox_Butterfat ~ Breed, data = data)

# Summarize the model
summary(model)

# Assumption 1: Linearity
# Plot residuals vs fitted values (visual inspection for patterns)
par(mfrow = c(2, 2))  # Set plotting layout
plot(model)

# Assumption 2: Homoscedasticity (constant variance)
# Using the Breusch-Pagan test
bptest(model)  # Test for heteroscedasticity

# Additional visual check: Scale-Location plot
# (already generated in `plot(model)` above)

# Assumption 3: No Autocorrelation
# Durbin-Watson test for autocorrelation of residuals
dwtest(model)

# Assumption 4: Normality of Residuals
# Q-Q plot for visual assessment of normality
qqnorm(residuals(model))
qqline(residuals(model), col = "red")

# Shapiro-Wilk test for normality
shapiro.test(residuals(model))
```
