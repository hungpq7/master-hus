In Sect. 2.7, an interpretation for a model with systematic component $ \mu = E[\log y] = \beta_0 + \beta_1x$ was discussed.
1. Use a Taylor series expansion of $\log y$ about $\mu = E[y]$.
2. Find the expected value of both sides of this equation, and hence show
that $E[\log y] ‚âà \log E[y] = \log \mu$.
3. Using this information, show that an increase in the value of $x$ by 1 is

---
Trong ph·∫ßn 2.7, m·ªôt c√°ch di·ªÖn gi·∫£i cho m√¥ h√¨nh v·ªõi th√†nh ph·∫ßn h·ªá th·ªëng $ \mu=E[log y]=\beta_0 + \beta_1x$ ƒë√£ ƒë∆∞·ª£c th·∫£o lu·∫≠n.
1. S·ª≠ d·ª•ng khai tri·ªÉn chu·ªói Taylor c·ªßa $\log y$ v·ªÅ $\mu = E[y]$.
2. T√¨m gi√° tr·ªã k·ª≥ v·ªçng c·ªßa c·∫£ hai v·∫ø c·ªßa ph∆∞∆°ng tr√¨nh n√†y, v√† t·ª´ ƒë√≥ ch·ª©ng minh r·∫±ng $E[\log y] ‚âà \log E[y] = \log \mu$.

## ***Solution 2.11.1. S·ª≠ d·ª•ng khai tri·ªÉn chu·ªói Taylor c·ªßa log y v·ªÅ Œº = E[y].***

C√¥ng th·ª©c khai tri·ªÉn Taylor c·ªßa m·ªôt h√†m $f(x)$ t·∫°i ƒëi·ªÉm $x_0$ ƒë∆∞·ª£c bi·ªÉu di·ªÖn nh∆∞ sau:


$f(x) = \sum_{n=0}^\infty \frac{f^{(n)}(x_0)}{n!} (x-x_0)^n$

<!-- > $= f(x_0) + \frac{f'(x_0)}{1!} (x-x_0) + \frac{f''(x_0)}{2!} (x-x_0)^2 + \frac{f^{(3)}(x_0)}{3!} (x-x_0)^3 + ...$ -->
> $= f(x_0) + f'(x_0)(x-x_0) + \frac{f''(a)}{2}(x-x_0)^2 + \frac{f^{(3)}(x_0)}{3!}(x-x_0)^3 + ...$

√Åp d·ª•ng khai tri·ªÉn Taylor v·ªõi h√†m  $\log y$ quanh $\mu$, ta c√≥:

$

$

>>> $= \log(\mu) + \frac{1}{\mu}(y - \mu) - \frac{1}{2\mu^2}(y - \mu)^2 + \dots $


## ***Solution 2.11.2. T√¨m gi√° tr·ªã k·ª≥ v·ªçng c·ªßa c·∫£ hai v·∫ø c·ªßa ph∆∞∆°ng tr√¨nh n√†y, v√† t·ª´ ƒë√≥ ch·ª©ng minh r·∫±ng $ùê∏[\logùë¶]‚âà\logùê∏[ùë¶]=\log\mu$.***

L·∫•y gi√° tr·ªã k·ª≥ v·ªçng \(E\) c·ªßa hai v·∫ø ph∆∞∆°ng tr√¨nh, ta c√≥:

> $E[\log(y)] = E[log(\mu)] + E[\frac{1}{\mu}(y-\mu)] - E[\frac{1}{2\mu^2}(y-\mu)^2] + ...$

Tri·ªÉn khai t·ª´ng th√†nh ph·∫ßn:
1. V√¨ $ \mu $ l√† h·∫±ng s·ªë, n√™n $ E[\log(\mu)] = \log(\mu) $.  
2. M·∫∑t kh√°c $E[y - \mu] = 0$, v√¨ $ \mu = E[y]$

>$=> E\left[\frac{1}{\mu}(y - \mu)\right] = \frac{1}{\mu}E[y - \mu] = 0 $  

3. H∆°n n·ªØa, $E[(y - \mu)^2] = var(y)$

>$=> E\left[\frac{-1}{2\mu^2}(y - \mu)^2\right] = \frac{-1}{2\mu^2}E[(y - \mu)^2] = \frac{-1}{2\mu^2} \text{var}(y) $.

K·∫øt qu·∫£:
> $E[\log(y)] = \log(\mu) - \frac{1}{2\mu^2} \text{var}(y) + \dots $

N√≥i c√°ch kh√°c:
> $E[\log(y)] \approx \log(\mu)$ (ƒëpcm)

Bi·∫øt r·∫±ng $E[\log(y)] \approx \log\mu \approx \beta_0 + \beta_1x$

ƒê·ªÉ xem x√©t s·ª± thay ƒë·ªïi c·ªßa $\mu$, ta l·∫•y $e$ m≈© cho c·∫£ hai v·∫ø:

> $e^{\log \mu} = \mu \approx e^{\beta_0 + \beta_1x}$

X√©t m√¥ h√¨nh tuy·∫øn t√≠nh:
$\mu = e^{\beta_0 + \beta_1x}$

Khi $x$ tƒÉng 1 ƒë∆°n v·ªã, ta c√≥:
>$\mu_{new} = e^{\beta_0 + \beta_1(x+1)} = e^{\beta_0 + \beta_1x} . e^{\beta_1} = \mu .e^{\beta_1}$

# Problem 2.12

2.12. Using R, produce a vector of 30 random numbers y from a standard normal distribution (use rnorm()). Generate a second vector of 30 random numbers x from a standard normal distribution. Find the P -value for testing if the explanatory variable x is significantly related to y using the regression model lm(y ~ x).

Repeat the process a large number of times, say 1000 times. What propor- tion of the P -values are less than 5%? Less than 10%? What is the lesson?

---

2.12. S·ª≠ d·ª•ng R, t·∫°o m·ªôt vector g·ªìm 30 s·ªë ng·∫´u nhi√™n y t·ª´ ph√¢n ph·ªëi chu·∫©n (s·ª≠ d·ª•ng h√†m rnorm()). T·∫°o m·ªôt vector th·ª© hai g·ªìm 30 s·ªë ng·∫´u nhi√™n x t·ª´ ph√¢n ph·ªëi chu·∫©n. T√¨m gi√° tr·ªã P -value ƒë·ªÉ ki·ªÉm tra xem bi·∫øn x c√≥ li√™n quan ƒë√°ng k·ªÉ ƒë·∫øn y hay kh√¥ng b·∫±ng c√°ch s·ª≠ d·ª•ng m√¥ h√¨nh h·ªìi quy lm(y ~ x).

L·∫∑p l·∫°i qu√° tr√¨nh n√†y nhi·ªÅu l·∫ßn, v√≠ d·ª• 1000 l·∫ßn. T√≠nh t·ª∑ l·ªá ph·∫ßn trƒÉm c√°c gi√° tr·ªã P-value nh·ªè h∆°n 5%? Nh·ªè h∆°n 10%? B√†i h·ªçc r√∫t ra l√† g√¨?


````r
# L·∫∑p l·∫°i 1000 l·∫ßn v√† t√≠nh t·ª∑ l·ªá P-value
set.seed(123)  # ƒê·∫£m b·∫£o k·∫øt qu·∫£ t√°i l·∫≠p
n_iterations <- 1000
p_values <- numeric(n_iterations)

for (i in 1:n_iterations) {
  y <- rnorm(30) # Vector y t·ª´ ph√¢n ph·ªëi chu·∫©n chu·∫©n
  x <- rnorm(30) # Vector x t·ª´ ph√¢n ph·ªëi chu·∫©n chu·∫©n
  model <- lm(y ~ x)
  p_values[i] <- summary(model)$coefficients[2, 4]
  #Ch·ªçn h√†ng th·ª© 2 trong b·∫£ng h·ªá s·ªë, t∆∞∆°ng ·ª©ng v·ªõi bi·∫øn x (bi·∫øn ƒë·ªôc l·∫≠p). Ch·ªçn c·ªôt th·ª© 4, t·ª©c l√† P-value cho h·ªá s·ªë h·ªìi quy c·ªßa bi·∫øn x.
}

# T√≠nh t·ª∑ l·ªá P-value nh·ªè h∆°n 5% v√† 10%
prop_less_5 <- mean(p_values < 0.05)
cat("T·ª∑ l·ªá P-value nh·ªè h∆°n 5%:", prop_less_5 * 100, "%\n")

prop_less_10 <- mean(p_values < 0.1)
cat("T·ª∑ l·ªá P-value nh·ªè h∆°n 10%:", prop_less_10 * 100, "%\n")
````
