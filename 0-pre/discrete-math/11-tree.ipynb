**14.** *Show that a simple graph is a tree if and only if it is connected but the deletion of any of its edges produces a graph that is not connected.*

- $A$: a simple graph is a tree
- $B$: a simple graph is connected but deletion of any edge produces a not-connected graph
- Prove $A\to B$: by the tree definition, there is exactly 1 path between any pair of vertices. So this graph must be connected in order to have at least 1 path between any pair of vertices. If we delete any edge, at least 1 path, let's say between pair of vertices $m$ and $n$ is discontinue. As there are no other $mn$ path, this graph is no longer a connected.
- Prove $B\to A$: a connected graph ensures there are at least 1 $mn$ path for any $m$ and $n$. This is necessary condition. If any edge deletion produces a connected graph, then an path $mn$ containing that edge must have a backup path, which violates tree definition. This is sufficient condition.
