### **Chapter 1: Probability and Random Variables**

#### **I. Probability Space**

A probability space is a triple $(\Omega, \mathcal{A}, P)$

* $\Omega$: sample space (set of all outcomes)


Within this:

1. $\Omega \neq \varnothing$ â€” called the sample space (set of all possible outcomes $\omega$)
2. $\mathcal{A}$: called the sigma-algebra â€” a collection of events (subsets of $\Omega$)

**Axioms of the sigma-algebra $\mathcal{A}$:**

* $\Omega, \varnothing \in \mathcal{A}$
* If $A \in \mathcal{A}$, then $A^c = \Omega \setminus A \in \mathcal{A}$
* If $A_i \in \mathcal{A}$, $i = 1, 2, \dots$, then $\bigcup_{i=1}^{\infty} A_i \in \mathcal{A}$

**Axiom (Probability Function):**

* $P: \mathcal{A} \to \mathbb{R}^+$ â€” $P(A)$ represents the probability of event $A$

**Probability Axioms:**

* $0 \leq P(A) \leq 1$
* If $A_i$ are mutually exclusive (disjoint events), then

  $$
  P\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} P(A_i)
  $$

#### **2. Event Space / Sigma-Algebra $\mathcal{A}$**

This is a collection of subsets of $\Omega$ (i.e. events), including:

* The empty set $\varnothing$ (impossible event),

* The full set $\Omega$ (certain event),

* Complements: If an event is in $\mathcal{A}$, its complement is too,

* Countable unions: If $A_1, A_2, \dots \in \mathcal{A}$, then $\bigcup A_i \in \mathcal{A}$.

* **In this case**:

  $$
  \mathcal{A} = \{\varnothing, \{1\}, \{2\}, \dots, \{6\}, \{1,2\}, \dots, \Omega\}
  $$

  (All $2^6 = 64$ subsets of $\Omega$ â€” the power set)

---

#### **3. Probability Measure $P$**

This assigns a number to each event in $\mathcal{A}$, representing how likely it is to occur.

* **In this case** (with a fair die):

  $$
  P(\{i\}) = \frac{1}{6} \quad \text{for each } i \in \{1,2,3,4,5,6\}
  $$

  And for any event (subset) $A \subseteq \Omega$:

  $$
  P(A) = \sum_{i \in A} P(\{i\}) = \frac{|A|}{6}
  $$

## 2.1. Definition
A random variable is naturally just a mapping from $\Omega$ to $\mathbb{R}$. Types:
- discrete: countable, limited values, enumerable $X(\Omega)={x_1,x_2,\dots}$. Frequency table: $P_i=P(X=x_i)$ represent PMF.

## 2.2. Expected value
- discrete: $E(X)=\sum{x_ip_i}$
- continuous: $E(X)=\int{xf(x)dx}$
- characteristic: some obvious ones
- consequence: proof of variance-mean relationship formula


### **Example**

Let $X$ be a random variable with probability density function (PDF):

$$
f(x) =
\begin{cases}
c e^{-x}, & \text{if } x \geq 0 \\
0, & \text{otherwise}
\end{cases}
$$

---

### **Questions:**

**(a)** Find the constant $c$

**(b)** Compute $\mathbb{E}[Y]$ where $Y = e^{-X}$

---

### **Solution:**

#### **(a)**

Since $f(x)$ is a PDF, it must satisfy:

1. $f(x) \geq 0$ â†’ always true if $c > 0$
2. $\int_{\mathbb{R}} f(x) dx = 1$

So:

$$
\int_{0}^{+\infty} c e^{-x} dx = 1
\Rightarrow c \left[-e^{-x} \right]_0^{+\infty} = 1
\Rightarrow c(0 + 1) = 1
\Rightarrow \boxed{c = 1}
$$

---

### **(b)** Compute $\mathbb{E}[Y]$, where $Y = e^{-X}$

We are given:

* $f_X(x) = e^{-x}$, for $x \ge 0$ (since $c = 1$)
* $Y = e^{-X}$ â†’ this is a transformation of a random variable.

#### âœ… **Method 1: Using expectation formula with transformation**

We compute the expectation directly:

$$
\mathbb{E}[Y] = \mathbb{E}[e^{-X}] = \int_0^{\infty} e^{-x} \cdot f_X(x) dx = \int_0^{\infty} e^{-x} \cdot e^{-x} dx = \int_0^{\infty} e^{-2x} dx
$$

$$
= \left[ \frac{-1}{2} e^{-2x} \right]_0^{\infty} = \left(0 + \frac{1}{2} \right) = \boxed{\frac{1}{2}}
$$

### âœ… Final Answers:

* (b) $\boxed{\mathbb{E}[Y] = \frac{1}{2}}$

## ðŸŽ© **Example (Vd2): Matching Hats Problem**

### **Problem Statement:**

There are $N$ people, each owns a unique hat.
At the end of an event, the hats are randomly shuffled and returned.
We ask:

---

### **(a) What is the probability that no one gets their own hat?**

(i.e., no fixed points in the permutation â€” called a *derangement*)

---

### **Step 1: Define Events**

Let:

* $A_i$: the event that person $i$ receives their own hat.
* $H$: the event that **at least one person** gets their own hat.

Then:

* $\bar{H}$: event that **no one** gets their own hat.
* We want to compute $P(\bar{H}) = 1 - P(H)$

---

### **Step 2: Inclusion-Exclusion Formula**

To compute $P(H) = P\left(\bigcup_{i=1}^{n} A_i\right)$, we use inclusion-exclusion:

$$
P\left(\bigcup_{i=1}^{n} A_i\right) = \sum_{i=1}^{n} P(A_i)
- \sum_{1 \le i < j \le n} P(A_i \cap A_j)
+ \sum_{1 \le i < j < k \le n} P(A_i \cap A_j \cap A_k)
- \dots + (-1)^{k+1} \sum P(A_{i_1} \cap \dots \cap A_{i_k}) + \dots
$$

---

### **Step 3: Compute Each Term**

Letâ€™s calculate $P(A_{i_1} \cap \dots \cap A_{i_k})$:

There are $(n - k)!$ permutations where $k$ people get their own hats.

So:

$$
P(A_{i_1} \cap \dots \cap A_{i_k}) = \frac{(n-k)!}{n!} = \frac{1}{k!}
$$

Also, the number of such $k$-combinations is $\binom{n}{k}$

So the total is:

$$
P(H) = \sum_{k=1}^{n} (-1)^{k+1} \frac{1}{k!}
\Rightarrow P(\bar{H}) = 1 - P(H) = \sum_{k=0}^{n} \frac{(-1)^k}{k!}
$$

---

### **Final Answer for (a):**

$$
\boxed{P(\text{no one gets their own hat}) = \sum_{k=0}^{n} \frac{(-1)^k}{k!}}
$$

As $n \to \infty$, this converges to:

$$
P(\bar{H}) \to \frac{1}{e} \approx 0.3679
$$

---

### **(b) What is the probability that at least one person gets their own hat?**

$$
P(H) = 1 - P(\bar{H}) = 1 - \sum_{k=0}^{n} \frac{(-1)^k}{k!}
\Rightarrow \boxed{P(\text{at least one correct}) = 1 - \frac{1}{e} \approx 0.6321}
$$
